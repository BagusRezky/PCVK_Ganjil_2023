# Langkah 1

## Import Library

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

# Langkah 2

## Membaca Image dari file KTP serta memberikan ukuran gambar x,y

```python
img_1 = cv.imread('./KTP.jpeg')
img_2 = cv.cvtColor(img_1, cv.COLOR_BGR2RGB)
cv2_imshow(img_1)
plt.imshow(img_2)
```

# Langkah 3

## Menutup bagian provinsi dan kabupaten seperti perintah dari soal!!

```python
img_4s = np.copy(img_1)
#tutup bagian provinsi
for y in range (210,610):
  for x in range (0,30):
    img_4s[x,y] = [110,0,255]

#tutup bagian Kabupaten
img_3 = np.copy(img_4s)
for y in range (250,560):
  for x in range (30,60):
    img_3[x,y] = [0,255,255]
final_frame = cv.hconcat((img_1, img_3))
cv2_imshow(final_frame)
```
